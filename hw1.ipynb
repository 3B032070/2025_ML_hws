```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
# 設定隨機seed
np.random.seed(42)
tf.random.set_seed(42)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def softmax(x):
    exp_x = np.exp(x - np.max(x))  # 防止溢出
    return exp_x / np.sum(exp_x)
```

# 輸入層

```python
X = np.random.rand(1, 6)  # 產生 6 維特徵值
print("X shape:", X.shape)
```

# 第一層：五個神經元

```python
W1 = np.random.rand(6, 5)
b1 = np.random.rand(5)
print("W1 shape:", W1.shape, "| Params:", np.prod(W1.shape))
print("b1 shape:", b1.shape, "| Params:", np.prod(b1.shape))

z1 = np.dot(X, W1) + b1
print("z1 shape:", z1.shape)
h1 = relu(z1)
print("h1 shape:", h1.shape)
```

# 第二層：四個神經元

```python
W2 = np.random.rand(5, 4)
b2 = np.random.rand(4)
print("W2 shape:", W2.shape, "| Params:", np.prod(W2.shape))
print("b2 shape:", b2.shape, "| Params:", np.prod(b2.shape))

z2 = np.dot(h1, W2) + b2
print("z2 shape:", z2.shape)
h2 = relu(z2)
print("h2 shape:", h2.shape)
```

# 第三層：四個神經元

```python

W3 = np.random.rand(4, 4)
b3 = np.random.rand(4)
print("W3 shape:", W3.shape, "| Params:", np.prod(W3.shape))
print("b3 shape:", b3.shape, "| Params:", np.prod(b3.shape))

z3 = np.dot(h2, W3) + b3
print("z3 shape:", z3.shape)
h3 = sigmoid(z3)
print("h3 shape:", h3.shape)
```

# 輸出層：三個神經元

```python
W4 = np.random.rand(4, 3)
b4 = np.random.rand(3)
print("W4 shape:", W4.shape, "| Params:", np.prod(W4.shape))
print("b4 shape:", b4.shape, "| Params:", np.prod(b4.shape))

z4 = np.dot(h3, W4) + b4
print("z4 shape:", z4.shape)
y_hat = softmax(z4)
print("y_hat shape:", y_hat.shape)
```

```python
# 計算 Cross-Entropy Error
y_true = np.array([[0, 0, 1]])  # 假設真實標籤
cross_entropy_error = -np.sum(y_true * np.log(y_hat))
print("Cross-Entropy Loss:", cross_entropy_error)
```

```python
# 視覺化 softmax 輸出
plt.bar(["Class 1", "Class 2", "Class 3"], y_hat[0])
plt.xlabel("Class")
plt.ylabel("Probability")
plt.title("Softmax Output Distribution")
plt.show()
```
